```python
from multiprocessing import Process, Queue, Value
import random
import time


def putItemIntoBuffer(item,buffer):
    buffer.put(item)
   
def productor(semaphore_emptyCount,semaphore_fillCount,buffer):
  while True:
    item = random.randint(1,100)
    print(f"Produjimos el item {item}")
    semaphore_emptyCount -= 1
    buffer.put(item)
    semaphore_fillCount += 1
      
def consumidor(semaphore_emptyCount,semaphore_fillCount,buffer):
  while True:
    semaphore_fillCount -= 1
    item = buffer.get()
    print(f"Los clientes consumieron el item {item}")
    semaphore_emptyCount += 1
    
    
def main():
  BUFFER_SIZE = 5
  semaphore_fillCount = 0; 
  semaphore_emptyCount = BUFFER_SIZE; 

  buffer = Queue(maxsize=semaphore_emptyCount)
    
  p = Process(target=productor, args=(semaphore_emptyCount,semaphore_fillCount,buffer))
  c = Process(target=consumidor, args=(semaphore_emptyCount,semaphore_fillCount,buffer))

  p.start()
  c.start()

  
    
main()
```
