```python
from multiprocessing import Process, Queue
from threading import Thread, Condition
import time
import random


# creamos la clase productor 
class Productor(Thread):
    def __init__(self, cola, condicion):
        Thread.__init__(self)
        self.condicion = condicion
        self.cola = cola
     
    def run(self):
        while True:
            lista = random.randint(0,100)
            # El productor adquiere el bloqueo
            self.condicion.acquire()
            print('Condicion adquirida por %s', self.name)
            # Agregamos a la cola el número generado por la lista
            self.cola.put(lista)
            # indicamos quién se lo agregó a la cola
            print('agregamos a la cola por %s', self.name)
            # le notificamos al consumidor que hay algo para recuperar
            print('Condicion notificada por %s',self.name)
            self.condicion.notify()
            print('Condicion liberada por %s', self.name)
            # liberamos el bloqueo y esperamos un segundo
            self.condicion.release()
            time.sleep(1)
 
# Creamos la clase consumidor
class Consumidor(Thread):
    def __init__(self,cola,condicion):
        Thread.__init__(self)
        self.condicion = condicion
        self.cola = cola
     
    def run(self):
        while True:
            # obtenemos el bloqueo
            self.condicion.acquire()
            print('condicion adquirida por %s', self.name)
            
            while True:
                # si la lista tiene elementos
                if self.cola:
                    # Usamos elemento de la cola y lo obtenemos(sacamos) 
                    
                    print('obtenemos el primer elemento de la cola y lo sacamos, hecho por %s', self.name)
                    self.cola.get()
                    break
                # si no hay nada, esperamos a que el producto lo notifique
                
                print('Condicion espera por %s', self.name)
                # libera el bloqueo para que el productor pueda adquirir el recurso y hacer su trabajo. 
                self.condicion.wait()
            # cuando el elemento se recupera de la cola, liberamos el bloqueo
            print('condicion liberada por %s', self.name)
            self.condicion.release()




cola = Queue()
cola.put([5,2,55,4,3,6,87,23,34,17])
condicion = Condition()
hilo1 = Productor(cola, condicion)
hilo2 = Consumidor(cola, condicion)
hilo1.start()
hilo2.start()
hilo1.join()
hilo2.join()
 
```
