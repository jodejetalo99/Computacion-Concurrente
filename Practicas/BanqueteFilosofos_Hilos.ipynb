```python
import threading

def filosofo(indice):
    print(f"Ya llegó el filósofo {indice}")
    while True:
        levanta_palillos(indice)
        come(indice)
        suelta_palillos(indice)

def levanta_palillos(indice):
    # Zurdo
    if indice == 0:
        # Adquiere un semáforo.
        palillos[indice].acquire()
        print(f"{indice} - Tengo el palillo izquierdo" )
        palillos[(indice+1) % tenedores].acquire()
    # Diestro
    else:
        # Adquiere un semáforo.
        palillos[(indice+1) % tenedores].acquire()
        print(f"{indice} - Tengo el palillo derecho")
        palillos[indice].acquire()
    print(f"{indice} - Tengo ambos palillos")

def suelta_palillos(indice):
    # Soltamos los semáforo, incrementando el contador interno en uno.
    palillos[(indice+1) % len(palillos)].release()
    palillos[indice].release()

def come(indice):
    print(f"El filósofo {indice} está comiendo")
   
# Número de tenedores
tenedores = 5

# Semáforos para la comunicación entre hilos
# Los palillos son un arreglo de semáforos
palillos = [threading.Semaphore(1) for i in range(tenedores)]

# Creamos los
filosofos = [threading.Thread(target=filosofo,args=[i]).start() for i in range(tenedores)]
```
