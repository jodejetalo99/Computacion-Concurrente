```python
from threading import Thread, Lock
import random 
import time
```

```python
class Tortuga(Thread):
    def __init__(self):
        self.contador_carrera = 1
        super().__init__()
    

    
    def run(self):
        
        print("Inicio de la carrera para la tortuga!")
        
        while self.contador_carrera < 70:

            prob = random.randint(1,100)
            
            if prob >= 1 and prob <=50:
                time
                self.contador_carrera += 3
                time.sleep(1)

            elif prob > 50 and prob <= 70:
                self.contador_carrera -= 6
                if self.contador_carrera < 1:
                    self.contador_carrera = 1
                time.sleep(1)

            else:
                self.contador_carrera += 1
                time.sleep(1)
            
            
            print(f"Posición-{self.contador_carrera} T: " + self.contador_carrera*" " + "T")
            
    

        
class Liebre(Thread):
    def __init__(self):
        self.contador_carrera = 1
        super().__init__()

    
    def run(self):
        
        print("Inicio de la carrera para la liebre!")
            
        while self.contador_carrera < 70:

            prob = random.randint(1,100)
            
            if prob >= 1 and prob <=20:
                self.contador_carrera += 0
                time.sleep(1)

            elif prob > 21 and prob <= 40:
                self.contador_carrera += 9
                time.sleep(1)

            elif prob > 41 and prob <= 50:
                self.contador_carrera -= 12
                if self.contador_carrera < 1:
                    self.contador_carrera = 1
                time.sleep(1)

            elif prob > 51 and prob <= 80:
                self.contador_carrera += 1
                time.sleep(1)
                
            else:
                self.contador_carrera -= 2
                if self.contador_carrera < 1:
                    self.contador_carrera = 1
                time.sleep(1)
            print(f"Posición-{self.contador_carrera} L: " + self.contador_carrera*" " + "L")
```

```python
tortuga = Tortuga()
liebre = Liebre()

tortuga.start()
liebre.start()

while True:
    if liebre.contador_carrera == 70 and tortuga.contador_carrera == 70:
        print("Empate")
        tortuga.join()
        liebre.join()
        break
        
    if liebre.contador_carrera >= 70:
        print("Ganó la liebre! Felicidades liebre!")
        print("Ánimo tortuga, termina la carrera por dignidad")
        liebre.join()
        tortuga.join()
        break
        
    if tortuga.contador_carrera >= 70:
        print("Ganó la tortuga! Felicidades tortuga!")
        print("Ánimo liebre, termina la carrera por dignidad")
        tortuga.join()
        liebre.join()
        break
```

```python
import numpy as np
import time

#  función que calcula el producto de dos matrices de manera secuencial:
def mult_matriz_sec(A,B):
    t = time.time()
    AB = np.array([[0 for k in range(len(B[0]))] for j in range(len(A))])

    for i,renglon in enumerate(A):

        for j,columna in enumerate([list(c) for c in zip(*B)]):

            AB[i][j] = sum([a*b for a,b in zip(renglon,columna)])
    print("Tiempo de ejecución: " +str(time.time()-t))
    return AB



# las pŕoximas cuatro funciones calculan el producto por cuadrantes
# c1 es el cuadrante 1, c2 es el cuadrante 2, c3 es el cuadrante 3 y c4 es el cuadrante 4
def mult_matriz_hilos_c1(A,B):
    global AB

    b = np.ceil(np.size(A,0)/2)
    a = np.ceil(np.size(B,1)/2)
    
    for j in range(0,b):
        for i in range(0,a):
            for k in range(np.size(B,1)):
                AB[j][i] += A[j][k]*B[k][i]
    
def mult_matriz_hilos_c2(A,B):
    global AB
    AB = np.zeros((np.size(B,1),np.size(A,0)))
    b = np.ceil(np.size(A,0)/2)
    a = np.ceil(np.size(B,1)/2)
    
    for j in range(b,b+b)):
        for i in range(0,a):
            for k in range(np.size(B,1)):
                AB[j][i] += A[j][k]*B[k][i]



    
def mult_matriz_hilos_c3(A,B):
    global AB
    b = np.ceil(np.size(A,0)/2)
    a = np.ceil(np.size(B,1)/2)
    
    for j in range(0,b):
        for i in range(a,a+a):
            for k in range(np.size(B,1)):
                AB[j][i] += A[j][k]*B[k][i]

    
def mult_matriz_hilos_c4(A,B):
    global AB
    b = np.ceil(np.size(A,0)/2)
    a = np.ceil(np.size(B,1)/2)
    
    for j in range(b*1,b+(1*b)):
        for i in range(a*1,a+(a*1)):
            for k in range(np.size(B,1)):
                AB[j][i] += A[j][k]*B[k][i]

    
    
```
